### Ejercicio 1
Utiliza `plotly` para realizar un  diagramas de torta sobre el estado conyugal de los clientes (resaltando el más predominante).

```python
# YOUR CODE HERE
Estado_Civil = df.groupby(by = "Marital_Status")["Marital_Status"].count()
values = []
for i in Estado_Civil:
    values.append(i)

fig = go.Figure(data = [go.Pie(labels = Estado_Civil.index, values = values, pull = [0, 0.03, 0, 0])])
fig.show()
```

### Ejercicio 2
Utiliza `plotly` para realizar un gráfico que permita verificar si el cambio en el monto de transacciones (cuarto trimestre a primer trimestre) se ajusta a una distribución normal.

```python
# YOUR CODE HERE
fig = go.Figure()

fig.add_trace(
    go.Histogram(
        x = list(df['Total_Amt_Chng_Q4_Q1']),
        y = list(range(len(df['Total_Amt_Chng_Q4_Q1']))),
        name = "Cambio en el monto de transacciones (4T a 1T)"
    )
)

fig.update_layout(
    title = "Cambio en el monto de transacciones (4T a 1T)",
    xaxis_title = "Numero de Cliente",
    yaxis_title = "Monto de las transacciones",

)

fig.show()
```

### Ejercicio 3
Utiliza `plotly` para realizar un gráfico de dispersión que permita visualizar la relación que existe entre las variables 'Year_Birth' y 'MntWines'.

```python
# YOUR CODE HERE

X = list(mercadeo['Year_Birth'])
Y = list(mercadeo['MntWines'])
```

```python
# YOUR CODE HERE
data = [go.Scatter(x = X,
                   y = Y,
                   mode = 'markers',
                   marker = dict(size = 12,
                                 color = 'rgb(255,0,0)',
                                 symbol = 'circle',
                                 line = {'width':2}
                                 ))]

layout = go.Layout(title = 'Year Birth vs. Mnt Wines',
                   xaxis = {'title':'Year Birth'},
                   yaxis = dict(title='Mnt Wines'),
                   hovermode = 'closest')

fig = go.Figure(data = data, layout=layout)

fig.show()
```

### Ejercicio 4
Utiliza `ipywidgets` para filtrar las observaciones de la base de datos de acuerdo con los siguentes criterios:
1. Nivel máximo de escolaridad alcanzado
2. Número de niños en el hogar
3. Número de adolescentes en el hogar

Asimismo, el 

```python
# YOUR CODE HERE
# creacio de funcion para usar como filtro 

def valores_filtro_columna(Col):
    Col = Col.astype(str)
    values = Col.unique().tolist()
    values = sorted(values)
    values.insert(0,"Todos")
    return values


def filtrar(Education, Kidhome, Teenhome):
    # Guardamos las columnas de interés en un nuevo DataFrame.
    DF = mercadeo[['Education','Kidhome','Teenhome']]
    # Los siguientes cuatro casos ajustan los registros a mostrar de acuerdo con la selección del usuario.
    if  Education != "Todos" and Kidhome != "Todos":
        display(DF[(DF["Education"] == Education) & (DF["Kidhome"] == Kidhome)
                   & (DF["Teenhome"] <= Teenhome)])
    elif Education == "Todos":
        display(DF[(DF["Kidhome"] == Kidhome) & (DF["Teenhome"] <= Teenhome)])
    elif Kidhome == "Todos":
        display(DF[(DF["Education"] == Education) & (DF["Teenhome"] <= Teenhome)])
    else:
        display(DF[(DF["Teenhome"] <= Teenhome)])
        

#Funcion Interactive

interactive_plot = interactive(
                                filtrar, 
                                Education = valores_filtro_columna(mercadeo["Education"]),
                                Kidhome = (0, 3, 1),
                                Teenhome = (0, 3, 1)
                              )

display(interactive_plot)
```

### Ejercicio 5
Utiliza `ipywidgets` y `plotly` para generar un gráfico con controles que:
1. Cuente con un control para seleccionar una de las variable de Gastos frente a los ingresos.
1. Genere un gráfico de dispersión
2. Genere un diagrama de caja.

```python
# YOUR CODE HERE
DF2 = mercadeo[['Income','MntWines','MntFruits','MntMeatProducts', 'MntFishProducts','MntSweetProducts','MntGoldProds']]

#crear barra
menu = widgets.Dropdown(
    options = ['MntWines','MntFruits','MntMeatProducts', 'MntFishProducts','MntSweetProducts','MntGoldProds'],
    value = 'MntWines',
    ddescription = "Variable: ",
    disabled=False,
)

seleccion = widgets.Output()

#Funcion para grafico de dispersion

def graficar(variable_independiente, variable_dependiente):
    data=[go.Scatter(x=DF2[variable_independiente], 
        y=DF2[variable_dependiente], 
        mode='markers', 
        marker= dict(size= 12, 
                      color= 'rgb(51,204,153)', 
                      symbol= 'circle', 
                      line= {'width':2} 
                      ) )] 
    
    layout=go.Layout(title= str(variable_independiente)+' vs. '+ str(variable_dependiente), 
                  xaxis= {'title':str(variable_independiente)}  , 
                  yaxis= dict(title=str(variable_dependiente)), 
                  hovermode= 'closest' ) 
    
    fig= go.Figure(data=data , layout=layout) 

    fig.show() 

#Diagrama de Caja
def diagrama_caja(variable_independiente, variable_dependiente):
    fig = go.Figure(
                  data = [ 
                      # Tipo de visualización. En este caso definimos un diagrama de caja.
                      go.Box(
                              y = DF2[variable_independiente], # Definimos los valores de la serie de datos.
                              name = "Income",                       # Definimos nombre de la serie de datos.
                              marker_color = 'yellow'                 # Definimos el color del diagrama.
                                ),
                      # Tipo de visualización. En este caso definimos un diagrama de caja.
                      go.Box(
                              y = DF2[variable_dependiente], # Definimos los valores de la serie de datos.
                              name = "Variable",                        # Definimos nombre de la serie de datos.
                              marker_color = 'green'                   # Definimos color del diagrama.
                                ) 
                        ],
                  # Modificamos el estilo del diseño.
                  layout = go.Layout(
                                      title = "variable de Gastos frente a los ingresos",  # Definimos el título del gráfico.
                                      xaxis_title = "Grupos",          # Definimos el título del eje x.
                                      yaxis_title = "Valor"            # Definimos el título del eje y.
                                    )
                  )

    fig.show()
    
def accion_cambio_menu(change):
    seleccion.clear_output() 
    with seleccion:
        if (change["new"] == "MntWines"):
            graficar('Income', "MntWines"),
            diagrama_caja('Income',"MntWines")

        elif (change["new"] == "MntFruits"):
            graficar('Income', "MntFruits"),
            diagrama_caja('Income', "MntFruits")

        elif (change["new"] == "MntMeatProducts"):
            graficar('Income', "MntMeatProducts"),
            diagrama_caja('Income', "MntMeatProducts")

        elif (change["new"] == "MntFishProducts"):
            graficar('Income', "MntFishProducts"),
            diagrama_caja('Income', "MntFishProducts")

        elif (change["new"] == "MntSweetProducts"):
            graficar('Income', "MntSweetProducts"),
            diagrama_caja('Income', "MntSweetProducts")

        elif (change["new"] == "MntGoldProds"):
            graficar('Income', "MntGoldProds"),
            diagrama_caja('Income', "MntGoldProds")

            
menu.observe(accion_cambio_menu, names='value')
display(menu, seleccion)


```
